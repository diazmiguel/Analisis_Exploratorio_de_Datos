```python
Cdf_ultimo_salario_mensual_o_retiro.plot()
q = 1000000
p = Cdf_ultimo_salario_mensual_o_retiro(q)
plt.vlines(
    x=q,
    ymin=0,
    ymax=p,
    color='black',
    linestyle='dashed'
)
plt.hlines(
    y=p,
    xmin=Pmf_ultimo_salario_mensual_o_retiro.qs[0],
    xmax=q,
    color='black',
    linestyle='dashed'
)

plt.plot(q,p, 'ro');
```

```python
Cdf_ultimo_salario_mensual_o_retiro.step()
p1=0.25
p2=0.75

ps=(0.25, 0.75)
qs= Cdf_ultimo_salario_mensual_o_retiro.inverse(ps)#regresa el valor inverso de lo preguntado

plt.vlines(
    x=qs,
    ymin=0,
    ymax=ps,
    color='black',
    linestyle='dashed'
)
plt.hlines(
    y=ps,
    xmin=Cdf_ultimo_salario_mensual_o_retiro.qs[0],
    xmax=qs,
    color='black',
    linestyle='dashed'
)
plt.scatter(
    x=qs,
    y=ps,
    color='red',
    zorder=2
);
```

```python
# LEY DE LOS GRANDES NUMEROS Y EL TEOREMA DEL LIMITE CENTRAL
#la probabilidad experimental tiende a la teorica a grandes muestras 
dado= empiricaldist.Pmf.from_seq([1,2,3,4,5,6])
dado.bar()
```

```python
for muestras in (1e2, 1e3, 1e4): #1e2 = 100 y asi con los valores
    muestras = int(muestras)
    valores = dado.sample(muestras)
    muestrasPmf = empiricaldist.Pmf.from_seq(valores)
    
    plt.figure(figsize=(5,5))
    muestrasPmf.bar()
    plt.grid()
    plt.axhline(y=1/6, color='red', linestyle='dashed')
    plt.ylim([0,0.50])
    plt.title(f" Tamaño de la muestra: {muestras}")
```

```python
dedicaciones=datos.dedicacion.replace(['Full-Time','Part-Time'],[1,0])
```

```python
numeroEjemplo=100
muestrasTomadas=30

muestritas = pd.DataFrame()

np.random.seed(42)
for i in range(1, numeroEjemplo + 1):
    muestrasTomadasEjemplo = dedicaciones.sample(muestrasTomadas, replace=True).to_numpy()
    muestrasNombre= f"ejemplo_{i}"
    muestritas[muestrasNombre]= muestrasTomadasEjemplo

mediaDedicacionFull = muestritas.mean().mean()
print(f"Estimacion de dedicacion Full-time en la poblacion es: {mediaDedicacionFull * 100 :.4f}%")
```

```python
muestrasMediaBinomial = pd.DataFrame(muestritas.mean(), columns=['sample_mean'])
sns.kdeplot(data=muestrasMediaBinomial)
plt.axvline(x=dedicaciones.mean(), color='red', linestyle='dashed');
```

```python
sample_size_experiment = pd.DataFrame(
    [[i, muestritas.iloc[:, 0:i].mean().mean().mean()] for i in range(1, numeroEjemplo + 1)],
    columns=['sample_size', 'estimated_mean']
)
```

```python
#grafica  del efecto
sns.scatterplot(
    data=sample_size_experiment,
    x='sample_size',
    y='estimated_mean'
)

plt.axhline(
    y=dedicaciones.mean(),
    color='red',
    linestyle='dashed'
)

plt.ylim([dedicaciones.mean() - 0.20, dedicaciones.mean() + 0.20])
```
